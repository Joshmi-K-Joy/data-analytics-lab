```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the file
covid_df = pd.read_csv("covid19_Confirmed_dataset.csv")
covid_df.head()
```

```python
covid_df.info()
```

```python
covid_df.describe()
```

```python
covid_df.shape
```

```python
corona_dataset_aggregated = covid_df.drop(['Lat', 'Long', 'Province/State'], axis=1)
corona_dataset_aggregated = corona_dataset_aggregated.groupby('Country/Region').sum()
corona_dataset_aggregated.head()
```

```python
corona_dataset_aggregated.loc["China"].plot(figsize=(10, 4))
plt.title("COVID-19 Growth Trend in China")
plt.xlabel("Date")
plt.ylabel("Confirmed Cases")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_5 = corona_dataset_aggregated.iloc[:, -1].sort_values(ascending=False).head(5)
top_5.plot(kind='bar', figsize=(8, 4), color='tomato', title="Top 5 Countries by Total Cases on 4/30/20")
plt.ylabel("Confirmed Cases")
plt.grid(True)
plt.show()
```

```python
transposed_df = corona_dataset_aggregated.T  # Dates become index
transposed_df.index = pd.to_datetime(transposed_df.index)  # convert to datetime
transposed_df[['India', 'US', 'Italy']].plot(figsize=(10, 5))
plt.title("COVID-19 Trend: India, US, Italy")
plt.xlabel("Date")
plt.ylabel("Confirmed Cases")
plt.grid(True)
plt.show()
```

## World happiness

```python
happiness_df = pd.read_csv("worldwide_happiness_report.csv")  # Replace with actual file name
happiness_df.head()
```

```python
# Drop non-numeric columns like 'Country' and 'Overall rank'
numeric_df = happiness_df.drop(['Country or region', 'Overall rank'], axis=1)

# Now plot the heatmap
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='YlGnBu')
plt.title('Correlation Heatmap of Happiness Factors')
plt.show()
```

```python
# Top 10 happiest countries based on Score
top_10 = happiness_df.sort_values(by='Score', ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='Score', y='Country or region', data=top_10, palette='viridis')
plt.title('Top 10 Happiest Countries (2020)')
plt.xlabel('Happiness Score')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=happiness_df, x='GDP per capita', y='Healthy life expectancy', hue='Score', palette='coolwarm', size='Score', sizes=(50, 200))
plt.title('GDP vs Life Expectancy (Color by Happiness Score)')
plt.xlabel('GDP per Capita')
plt.ylabel('Healthy Life Expectancy')
plt.tight_layout()
plt.show()
```

```python
# Rename for merging
covid_df_agg = covid_df.groupby("Country/Region").sum().drop(columns=['Lat', 'Long'])

# Latest total COVID cases per country (last column)
covid_latest = covid_df_agg.iloc[:, -1].reset_index()
covid_latest.columns = ['Country', 'Total_Cases']

# Merge with happiness data
merged_df = pd.merge(happiness_df, covid_latest, left_on='Country or region', right_on='Country', how='inner')

# Plot Happiness Score vs Total COVID Cases
plt.figure(figsize=(10,6))
sns.scatterplot(data=merged_df, x='Score', y='Total_Cases', hue='Country', legend=False)
plt.title('Happiness Score vs COVID-19 Total Cases')
plt.xlabel('Happiness Score')
plt.ylabel('Total COVID-19 Cases')
plt.yscale('log')  # Optional: log scale to handle wide range
plt.tight_layout()
plt.show()
```
