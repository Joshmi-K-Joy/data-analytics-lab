Assignment 1

 Step 1: Import NumPy

```python
import numpy as np
```

1. Create a 1D NumPy array from 1 to 20


a. Sum, Mean, Median, Std Dev

```python
arr_1d = np.arange(1, 21)
print("Sum:", np.sum(arr_1d))
print("Mean:", np.mean(arr_1d))
print("Median:", np.median(arr_1d))
print("Standard Deviation:", np.std(arr_1d))
```

b. Indices of elements > 10

```python
indices_gt_10 = np.where(arr_1d > 10)[0]
print("Indices of elements > 10:", indices_gt_10)
```

 2. 2D Array (4x4) with numbers from 1 to 16

```python
arr_2d = np.arange(1, 17).reshape(4, 4)
print("Original Array:\n", arr_2d)
```

a. Transpose the array

```python
print("Transpose:\n", arr_2d.T)
```

b. Row-wise and column-wise sum

```python
print("Row-wise Sum:", np.sum(arr_2d, axis=1))
print("Column-wise Sum:", np.sum(arr_2d, axis=0))
```

3. Two random 3x3 arrays from 1 to 20

```python
arr_a = np.random.randint(1, 21, size=(3, 3))
arr_b = np.random.randint(1, 21, size=(3, 3))

print("Array A:\n", arr_a)
print("Array B:\n", arr_b)
```

a. Element-wise operations

```python
print("Addition:\n", arr_a + arr_b)
print("Subtraction:\n", arr_a - arr_b)
print("Multiplication:\n", arr_a * arr_b)
```

b. Dot product

```python
print("Dot Product:\n", np.dot(arr_a, arr_b))
```

4. Reshape and Slice

```python
arr_reshape = np.arange(1, 13).reshape(3, 4)
print("Reshaped Array:\n", arr_reshape)
print("First 2 Rows & Last 2 Columns:\n", arr_reshape[:2, -2:])
```

Assignment 2

 Step 1: Import Pandas

```python
import pandas as pd
```

1. Create a DataFrame

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}

df = pd.DataFrame(data)
df.head()  # a. Show first 5 rows
```

2. Summary Statistics (Age, Salary)

```python
df[['Age', 'Salary']].describe()  # b. Summary of age and salary
```

 3. Average Salary of HR Employees

```python
df[df['Department'] == 'HR']['Salary'].mean()
```

4. Add Bonus Column (10% of Salary)

```python
df['Bonus'] = df['Salary'] * 0.10
df
```

5. Filter Employees Aged Between 25 and 30

```python
df[(df['Age'] >= 25) & (df['Age'] <= 30)]
```

6. Group by Department & Average Salary

```python
df.groupby('Department')['Salary'].mean().reset_index()
```

 7. Sort by Salary and Save to CSV

```python
sorted_df = df.sort_values(by='Salary', ascending=True)
sorted_df.to_csv("sorted_employees.csv", index=False)
sorted_df
```

Assignment 3

Step 1: Import matplotlib and numpy libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

1. Line Plot

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

plt.figure(figsize=(8, 4))
plt.plot(x, y, marker='o', linestyle='--', color='blue')
plt.title("Line Plot Example")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

2. Bar Graph  Student Marks

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

plt.figure(figsize=(8, 4))
plt.bar(students, marks, color=['purple', 'green', 'blue', 'orange'])
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

3. Pie Chart  Revenue Distribution

```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

explode = [0.1 if val == max(revenue) else 0 for val in revenue]

plt.figure(figsize=(6, 6))
plt.pie(revenue, labels=regions, autopct='%1.1f%%', startangle=90, explode=explode)
plt.title("Revenue Distribution by Region")
plt.axis('equal')
plt.show()
```

 4. Histogram  Random Integers from 1 to 100

```python
data = np.random.randint(1, 101, size=1000)

plt.figure(figsize=(8, 4))
plt.hist(data, bins=20, color='skyblue', edgecolor='black')
plt.title("Histogram of Random Integers (1-100)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
