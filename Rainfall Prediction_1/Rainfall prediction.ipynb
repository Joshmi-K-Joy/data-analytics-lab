```python
#importing libraries
import numpy as np
import pandas as pd
import sklearn as sk
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("austin_weather_final.csv")
data
```

```python
X=data.drop(['PrecipitationSumInches'],axis=1)
```

```python
Y=data["PrecipitationSumInches"]
```

```python
#reshaping it into 2d vector
Y=Y.values.reshape(-1,1)
```

```python
Y
```

```python
day_index=798
days =[i for i in range(Y.size)]
```

```python
#initialize the linear regression classifier
clf=LinearRegression()

#train the classifier
clf.fit(X,Y)
```

```python
#Plot the graph
print("The Preception trend graph: ")
plt.scatter(days,Y,color='g')
plt.scatter(days[day_index], Y[day_index],color='r')
plt.title("Preception level")
plt.xlabel("Days")
plt.ylabel("Precepitation in inches")
plt.show()
x_axis=X.filter(["TempAvgF","DewPointAvgF","HumidityAvgPercent","SeaLevelPressureAvgInches","VisibilityHighMiles","WindAvgMPH"])
```

```python
print("The Precipitation vs attributes trend graph: ")

for i in range(x_axis.columns.size):
    plt.subplot(3,2,i+1)
    plt.scatter(days,x_axis[x_axis.columns.values[i][:50]],color='g')
    plt.title(x_axis.columns.values[i])
plt.show()
```
